```python
# сортирую, чтобы было удобнее группировать сессии
df = df.sort_values(by = ['session_id', 'event_time'])
df
```

```python
gk = df[['session_id', 'event_time', 'event_name']]
gk = gk.sort_values(by = ['session_id', 'event_time'])
pd.set_option('display.max_rows', 500)
gk.head(100)
```

```python
# группирую по сессии, оставляя первое действие
f = df.groupby('session_id').agg(lambda x: x.iloc[0])
f.sort_values(by = ['session_id', 'event_time'])
f
```
